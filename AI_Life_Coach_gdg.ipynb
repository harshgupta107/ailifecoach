```python
# ── Cell 3: Set Your Goals ──

goals = []

print('=' * 45)
print('        🎯 SET YOUR GOALS')
print('=' * 45)
print('Categories: health, career, learning, mindset, finance')
print('Press Enter after adding each goal.')
print('Leave blank and press Enter when done.')
print('-' * 45)

while True:
    title = input('\nWhat is your goal? (blank = done): ').strip()
    if not title:
        break
    category = input('Category (health/career/learning/mindset/finance): ').strip() or 'general'
    why = input('Why is this goal important to you? ').strip()
    goals.append({'title': title, 'category': category, 'why': why})
    print(f'  ✅ Goal added: "{title}"')

print('\n' + '=' * 45)
if goals:
    print(f'  Total {len(goals)} goal(s) saved')
    for i, g in enumerate(goals, 1):
        print(f'  {i}. [{g["category"]}] {g["title"]}')
else:
    print('  No goals added. Run this cell again.')
print('=' * 45)
```

```python
# ── Cell 4: Daily Check-in ──

from datetime import datetime

checkins = []

print('=' * 45)
print("       📝 TODAY'S CHECK-IN")
print('=' * 45)

print('\nEnter your mood score (1-5):')
print('  1 = 😞 Very bad    2 = 😕 Not great')
print('  3 = 😐 Okay        4 = 🙂 Good')
print('  5 = 😄 Great')

while True:
    try:
        mood = int(input('\nMood (1-5): '))
        if 1 <= mood <= 5:
            break
        print('Please enter a number between 1 and 5!')
    except:
        print('Numbers only!')

wins       = input('\nWhat did you do / achieve today?\n> ').strip()
challenges = input('\nWhat was challenging today?\n> ').strip()

mood_emojis = {1:'😞', 2:'😕', 3:'😐', 4:'🙂', 5:'😄'}

checkin = {
    'date': datetime.now().strftime('%d %b %Y, %I:%M %p'),
    'mood': mood,
    'wins': wins,
    'challenges': challenges
}
checkins.append(checkin)

print('\n' + '=' * 45)
print(f'  ✅ Check-in saved! {checkin["date"]}')
print(f'  Mood: {mood_emojis[mood]} ({mood}/5)')
print('=' * 45)
```

```python
# ── Cell 5: AI Coach Chat ──

print('=' * 45)
print('       ✦ AI LIFE COACH CHAT')
print('=' * 45)
print('Ask your AI Coach anything!')
print('Type "quit" to end the chat.')
print('-' * 45)

# Build context
goals_text = '\n'.join([f"- {g['title']} ({g['category']}): {g['why']}" for g in goals]) if goals else 'No goals set yet.'

checkin_text = '\n'.join([
    f"- {c['date']} | Mood: {c['mood']}/5 | Wins: {c['wins']} | Challenges: {c['challenges']}"
    for c in checkins[-3:]
]) if checkins else 'No check-ins yet.'

system_prompt = f"""You are a warm, encouraging AI Life Coach.
Always reply in the same language the user writes in (Hindi, English, or Hinglish).
Be specific, actionable, and supportive. Keep replies to 3-5 sentences.

User's goals:
{goals_text}

Recent check-ins:
{checkin_text}
"""

chat = model.start_chat(history=[])
chat.send_message(system_prompt)

# Show quick prompts
print('\n💡 Quick prompts:')
print('  1. Analyze my progress')
print('  2. Give me motivation for today')
print('  3. What habits should I build?')
print('-' * 45)

while True:
    user_input = input('\nYou: ').strip()
    if not user_input:
        continue
    if user_input.lower() in ['quit', 'exit', 'bye']:
        print('\n✦ Coach: All the best! Keep going! 💪')
        break
    try:
        response = chat.send_message(user_input)
        print(f'\n✦ Coach: {response.text}')
        print('-' * 45)
    except Exception as e:
        print(f'❌ Error: {e}')
```

```python
# ── Cell 6 (Bonus): Summary Report ──

print('=' * 45)
print('         📊 YOUR SUMMARY')
print('=' * 45)

print(f'\n🎯 Total Goals: {len(goals)}')
for g in goals:
    print(f'   • [{g["category"]}] {g["title"]}')

print(f'\n📝 Total Check-ins: {len(checkins)}')
if checkins:
    avg_mood = sum(c['mood'] for c in checkins) / len(checkins)
    mood_emojis = {1:'😞', 2:'😕', 3:'😐', 4:'🙂', 5:'😄'}
    print(f'   Average Mood: {avg_mood:.1f}/5 {mood_emojis[round(avg_mood)]}')
    print(f'   Last Check-in: {checkins[-1]["date"]}')

if goals and checkins:
    summary_prompt = f"""
    Based on these goals: {goals_text}
    And these check-ins: {checkin_text}
    Give a 3-line motivational summary and one key action for tomorrow.
    Reply in same language as the check-in content.
    """
    response = model.generate_content(summary_prompt)
    print(f'\n✦ AI Summary:\n{response.text}')

print('\n' + '=' * 45)
print('       Keep going! You got this! 💪')
print('=' * 45)
```
